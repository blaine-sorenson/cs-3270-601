Assignment 4 - Words
==================================

There is a file `words.txt` that is found in the folder `hw4`.  This file contains 81520 words, one per line. You will use it for
this assignment as a lexicon of English words.    

## Program Description

Write a python program (external) in the `hw4` folder that chooses
a word at random from the lexicon and displays it to the user in a
random, scrambled order. Then let the user guess words that can be
formed with the letters of the word. 

After every guess, respond to the user whether the guess was
correct, and add it to the list of words that have been guessed 
so far. Print results grouped by length of words.

### Other Requirements

When running the first thing you are going to ask the user is the range of
word lengths to use.  

    Enter the range of word lengths (low,high): 

You will prompt the user for a guess, if the user enters a word that doesn't
fit the length of the word range, print error message `incorrect word size`
and ask again. (This is ignored if the entry is `q` as you will see below).   

    Enter a guess: thisistoolong
    The guess does not fit within the range of (low, high)
    
    Enter a guess: 

**Note: The low and high in the above error should be filled with the low and
high values supplied by the user.**   

When entering the string `“q”` it will quit the program.  

After every guess you will display a list of guessed words in alphabetical order
with blanks for non-guessed words using the format shown below.  

    ['air', 'art', 'tar', '---', '---', '---']
    3 3-letter words left
    ['tart']
    0 4-letter words left
    ['-----']
    1 5-letter words left
    
**Note: This is a demonstration of output when low is 3 and high is 5, but it will
need to be adjusted for different low/high values**   

When the program exists it should display the list of all the words.  

    3: ['air', 'art', 'rat', 'tar', 'tat', 'tit']
    4: ['tart']
    5: ['trait']
    
**Note**: If there are no words of a particular length, just ignore them altogether 
as shown below.   

Finally since this program is based on random-ness you will need to use the
`random` python package.  Be sure to add the ability for the random seed to
be supplied when the program is run, or default to None if not supplied.   

    esaywa:
    
    ['away', '----', 'ayes', 'easy', 'sway', 'ways', 'yaws', 'yeas', 'yews']
    1 4-letter words left
    ['------']
    1 6-letter words left

## Submission

For this assignment you will need to have your python program in the `hw4` folder and it
must be named `wordgame.py`.   

This can be executed like this:   

    python wordgame.py
    
It must also take in an optional argument which is the seed for the random number generator:   

    python wordgame.py 101010
    
After your program is complete commit it to github and create a pull request.
